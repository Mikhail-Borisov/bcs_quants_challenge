# Preview

```python
import numpy as np
import pandas as pd

import matplotlib.pyplot as plt
%matplotlib inline
```

load data for a given **label**

```python
label = 'RTSI'
timestamp = np.load('./data/m1/'+label+'/time.npy', encoding = 'bytes')
price = np.load('./data/m1/'+label+'/close.npy')
```

plot the data

```python
plt.rcParams['figure.figsize'] = (20.0, 10.0)
plt.plot(timestamp, price)
plt.title(label)
plt.ylabel('Price')
plt.xlabel('Date');
```
